### Question-1

You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

    10

   /   \

 2      7

/ \

8 4

Output:

    8

  /   \

4     10

/ \

2 7

```python
class TreeNodes:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inOrderTraversals(root, val):
    if root is None:
        return

    inOrderTraversals(root.left, val)
    val.append(root.value)
    inOrderTraversals(root.right, val)

def convertToBSt(root, val):
    if root is None:
        return

    convertToBSt(root.left, val)
    root.value = val.pop(0)
    convertToBSt(root.right, val)

def binaryTreeToBSt(root):
    val = []
    inOrderTraversals(root, val)
    val.sort()
    convertToBSt(root, val)

# Example usage
root = TreeNodes(10)
root.left = TreeNodes(2)
root.right = TreeNodes(7)
root.left.left = TreeNodes(8)
root.left.right = TreeNodes(4)

print("Input:")
print("      10")
print("    /  /")
print("   2  7")
print(" / /")
print("8 4")

binaryTreeToBSt(root)

print("\nOutput:")
print("   8")
print("  /  /")
print(" 4  10")
print("/ /")
print("2 7")

     
```

### Question-2:

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Consider the following BST:

Input-1:

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

Output-1:

The distance between the two keys = 4

Input-2:

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

Output-2:

The distance between the two keys = 2



```python
class TreeNodes:
    def __init__(self, value):
        self.value= value
        self.left = None
        self.right = None

def findLCa(root,node11,node12):
    if root.value > node11.value and root.value > node12.value:
        return findLCa(root.left, node11, node12)
    elif root.value < node11.value and root.value < node12.value:
        return findLCa(root.right, node11, node12)
    else:
        return root

def findDistances(root, target):
    if root.value == target.value:
        return 0
    elif root.value > target.value:
        return 1 + findDistances(root.left, target)
    else:
        return 1 + findDistances(root.right, target)

def findNodes(root, value):
    if root is None or root.value == value:
        return root
    elif root.value > value:
        return findNodes(root.left, value)
    else:
        return findNodes(root.right, value)

def findDistanceBetweenNodes(root, value11, value12):
    node11 = findNodes(root, value11)
    node12 = findNodes(root, value12)

    lca = findLCa(root, node11, node12)
    distance1 = findDistances(lca, node11)
    distance2 = findDistances(lca, node12)

    return distance1 + distance2

     
```

```python
root = TreeNodes(8)
root.left = TreeNodes(3)
root.right = TreeNodes(10)
root.left.left = TreeNodes(1)
root.left.right = TreeNodes(6)
root.left.right.left = TreeNodes(4)
root.left.right.right = TreeNodes(7)
root.right.right = TreeNodes(14)
root.right.right.left = TreeNodes(13)

values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
value11 = 6
value12 = 14
distance = findDistanceBetweenNodes(root, value11, value12)
print()
print("The distance between the two keys =", distance)
```

### Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

 10

  /   \

 5       20

    /   \

    30     35

Output:

5 10 30 20 35

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def convertToDLLUtils(root, previous):
    if root is None:
        return

    convertToDLLUtils(root.left, previous)

    root.left = previous[0]
    if previous[0] is not None:
        previous[0].right = root

    previous[0] = root

    convertToDLLUtils(root.right, previous)

def convertToDLl(root):
    previous = [None]
    convertToDLLUtils(root, previous)

def printDLl(head):
    current = head
    while current is not None:
        print(current.value, end=" ")
        current = current.right
    print()

     

root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

print("Input:")
print(" 10")
print(" /  /")
print(" 5 20")
print("   / /")
print("  30 35")

convertToDLl(root)

print("\nOutput:")
printDLl(root.left)
     
```

### Question-4:

Write a program to connect nodes at the same level.

Input:

    1

  /   \

2      3

/ \ / \

4 5 6 7

Output:

1 → -1

2 → 3

3 → -1

4 → 5

5 → 6

6 → 7

7 → -1



```python
class Nodes:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None

def connectNodes(root):
    if root is None:
        return

    level_started = root

    while level_started:
        current = level_started
        dummy = Node(0)
        previous = dummy

        while current:
            if current.left:
                previous.next = current.left
                previous = current.left

            if current.right:
                previous.next = current.right
                previous = current.right

            current = current.next

        level_started= dummy.next

def printConnectedNodes(root):
    current = root
    while current:
        print(current.value, end=" → ")
        current = current.next
    print("-1")
     

root = Nodes(1)
root.left = Nodes(2)
root.right = Nodes(3)
root.left.left = Nodes(4)
root.left.right = Nodes(5)
root.right.left = Nodes(6)
root.right.right = Nodes(7)

print("Input:")
print("  1")
print("/    /")
print("2   3")
print("//  / /")
print("4 5 6 7")

connectNodes(root)

print("\nOutput:")
printConnectedNodes(root)
     
```
